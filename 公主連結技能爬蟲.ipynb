```python
from selenium import webdriver
import time
from webdriver_manager.chrome import ChromeDriverManager

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument("--test-type")
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://pcrdfans.com/tools/timeline')
```

```python
myDict = {}
```

```python
for i in range(1,139):
    # click radio button
    python_button = driver.find_elements_by_xpath('//*[@id="__next"]/div[2]/div[2]/div[3]/div/div[2]/div[1]/div[' + str(i) + ']')[0]
    python_button.click()
    
    time.sleep(2)
    
    name = driver.find_elements_by_xpath('//*[@id="__next"]/div[2]/div[2]/div[3]/div/div[3]/div/div[2]/div/div[1]/h3')[0].text
    
    myDict[name] = {'Name':name}
    
    lens = len(driver.find_elements_by_class_name("ant-timeline-item"))
    skills = []
    for i in range(1,lens + 1):
        path = '//*[@id="__next"]/div[2]/div[2]/div[3]/div/div[3]/ul/li[ '+ str(i) + ' ]/div[3]'
        data = driver.find_elements_by_xpath(path)[0].text
        skills.append(data)
        
    myDict[name] = {'Skills':skills}
    
    getBack = driver.find_elements_by_xpath('//*[@id="__next"]/div[2]/div[2]/div[3]/div/div[1]/div[2]/button')[0]
    getBack.click()
    
    time.sleep(2)
```

```python
print(myDict)
```

```python
import pandas as pd
```

```python
names = []
for key, value in myDict.items():
    names.append(key)
```

```python
times = list(range(90,-1,-1))
for i in range(len(times)):
    times[i] = str(times[i] // 60) + ':' + str(times[i] % 60)
```

```python
timeData = pd.DataFrame(columns = times, index = names)
```

```python
for key, value in myDict.items():
    for i in range(len(myDict[key]['Skills'])):
        text = myDict[key]['Skills'][i]
        timeData.loc[key][text.split(' ')[0]] = text.split(' ')[1]
```

```python
timeData.to_csv('timeData.csv',encoding="utf_8_sig")
```

```python
import tkinter as tk
from tkinter import StringVar
from tkinter import ttk
from PIL import ImageTk, Image
from tkinter.filedialog import askdirectory
import os
from tinkerTest import process
```

```python
app = tk.Tk()
app.title('Princess')
app.geometry('1400x900')
app.resizable(0, 0)

count = 0
charList = []
select1 = None
select2 = None
select3 = None
select4 = None
select5 = None
def selectChar(app,file):
    global count,charList
    if file not in charList:
        if count == 5:
            return
        charList.append(file)
        count += 1
        new(app,count)
    else:
        charList.remove(file)
        count -= 1
        delete(count)
        edit(app,count)
    print(charList)
    return
def new(app,count):
    global select1,select2,select3,select4,select5
    if count == 1:
        imgPath = charList[0]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select1 = tk.Label(app, image = selectimg)
        select1.image = selectimg
        select1.place(x=40,y=350)
    elif count == 2:
        imgPath = charList[1]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select2 = tk.Label(app, image = selectimg)
        select2.image = selectimg
        select2.place(x=40,y=450)   
    elif count == 3:
        imgPath = charList[2]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select3 = tk.Label(app, image = selectimg)
        select3.image = selectimg
        select3.place(x=40,y=550)
    elif count == 4:
        imgPath = charList[3]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select4 = tk.Label(app, image = selectimg)
        select4.image = selectimg
        select4.place(x=40,y=650)
    else:
        imgPath = charList[4]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select5 = tk.Label(app, image = selectimg)
        select5.image = selectimg
        select5.place(x=40,y=750)
def edit(app,count):
    global select1,select2,select3,select4,select5
    print(count)
    if count >= 1:
        imgPath = charList[0]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select1 = tk.Label(app, image = selectimg)
        select1.image = selectimg
        select1.place(x=40,y=350)
    if count >= 2:
        imgPath = charList[1]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select2 = tk.Label(app, image = selectimg)
        select2.image = selectimg
        select2.place(x=40,y=450)   
    if count >= 3:
        imgPath = charList[2]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select3 = tk.Label(app, image = selectimg)
        select3.image = selectimg
        select3.place(x=40,y=550)
    if count >= 4:
        imgPath = charList[3]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select4 = tk.Label(app, image = selectimg)
        select4.image = selectimg
        select4.place(x=40,y=650)
    if count >= 5:
        imgPath = charList[4]
        image = Image.open(imgPath)
        image = image.resize((90, 90))
        selectimg = ImageTk.PhotoImage(image)
        select5 = tk.Label(app, image = selectimg)
        select5.image = selectimg
        select5.place(x=40,y=750)
        
def delete(count):
    global select1,select2,select3,select4,select5
    if select1:
        select1.configure(image = '')
    if select2:
        select2.configure(image = '')
    if select3:
        select3.configure(image = '')
    if select4:
        select4.configure(image = '')
    if select5:
        select5.configure(image = '')
    return

imgPath1 = './princess/hiyoriNewyear.PNG'
image = Image.open(imgPath1)
image = image.resize((60, 60))
char1img = ImageTk.PhotoImage(image)
char1 = tk.Button(app, image = char1img,command= lambda: selectChar(app,imgPath1))
char1.place(x=0,y=0)

imgPath2 = './princess/rei.PNG'
image = Image.open(imgPath2)
image = image.resize((60, 60))
char2img = ImageTk.PhotoImage(image)
char2 = tk.Button(app, image = char2img,command=lambda: selectChar(app,imgPath2))
char2.place(x=60,y=0)

imgPath3 = './princess/gitta.PNG'
image = Image.open(imgPath3)
image = image.resize((60, 60))
char3img = ImageTk.PhotoImage(image)
char3 = tk.Button(app, image = char3img,command=lambda: selectChar(app,imgPath3))
char3.place(x=120,y=0)

imgPath4 = './princess/christina.PNG'
image = Image.open(imgPath4)
image = image.resize((60, 60))
char4img = ImageTk.PhotoImage(image)
char4 = tk.Button(app, image = char4img,command=lambda: selectChar(app,imgPath4))
char4.place(x=180,y=0)


imgPath5 = './princess/shizuruvalentine.PNG'
image = Image.open(imgPath5)
image = image.resize((60, 60))
char5img = ImageTk.PhotoImage(image)
char5 = tk.Button(app, image = char5img,command=lambda: selectChar(app,imgPath5))
char5.place(x=240,y=0)

imgPath6 = './princess/line.PNG'
image = Image.open(imgPath6)
image = image.resize((60, 60))
char6img = ImageTk.PhotoImage(image)
char6 = tk.Button(app, image = char6img,command=lambda: selectChar(app,imgPath6))
char6.place(x=300,y=0)

enter = tk.Button(app,text="確定",borderwidth=3,relief="solid",height = 2,  width=10,bg="White")
enter.config(font=("Courier", 15))
enter.place(x=400,y= 40)

header = tk.Label(text="顯示破防堆疊狀態",borderwidth=3,relief="solid",height = 2,  width=73,bg="White")
header.config(font=("Courier", 20))
header.place(x=150,y= 240)


# imgPath = './princess/rei.PNG'

# image = Image.open(imgPath)
# image = image.resize((90, 90))
# oneimg = ImageTk.PhotoImage(image)
# one = tk.Label(app, image = oneimg)
# one.place(x=40,y= 350)

# imgPath = './princess/hiyoriNewyear.PNG'

# image = Image.open(imgPath)
# image = image.resize((90, 90))
# twoimg = ImageTk.PhotoImage(image)
# two = tk.Label(app, image = twoimg)
# two.place(x=40,y= 450)

# imgPath = './princess/gitta.PNG'

# image = Image.open(imgPath)
# image = image.resize((90, 90))
# threeimg = ImageTk.PhotoImage(image)
# three = tk.Label(app, image = threeimg)
# three.place(x=40,y= 550)


# imgPath = './princess/christina.PNG'

# image = Image.open(imgPath)
# image = image.resize((90, 90))
# fourimg = ImageTk.PhotoImage(image)
# four = tk.Label(app, image = fourimg)
# four.place(x=40,y= 650)


# imgPath = './princess/shizuruvalentine.PNG'

# image = Image.open(imgPath)
# image = image.resize((90, 90))
# fiveimg = ImageTk.PhotoImage(image)
# five = tk.Label(app, image = fiveimg)
# five.place(x=40,y= 750)

imgPath = './princess/combine.PNG'

image = Image.open(imgPath)
image = image.resize((1200, 500))
outputimg = ImageTk.PhotoImage(image)
output = tk.Label(app, image = outputimg)
output.place(x=150,y= 350)


app.mainloop()
```
